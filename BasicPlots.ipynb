```python
import matplotlib.pyplot as plt
```

```python
nice_colors = ['powderblue', 'palevioletred', 'palegoldenrod', 'darkgray']

# named colors: https://matplotlib.org/2.0.2/examples/color/named_colors.html 
```

```python
labels = ['S1', 'S2', 'S3', 'S4', 'S5']
market1 = [20, 35, 30, 35, 27]
market2 = [25, 32, 34, 20, 25]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, market1, width, label='Concrete', color='palevioletred')
ax.bar(labels, market2, width, bottom=market1, label='Charcrete', color='powderblue')

ax.set_ylabel('ylabel')
ax.set_xlabel('xlabel')
ax.set_title('Title')
ax.legend(loc="upper right")

plt.show()
```

```python
labels = 'Market1', 'Market2', 'Market3', 'Market4'
sizes = [15, 30, 45, 10]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, counterclock=False, colors = nice_colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
